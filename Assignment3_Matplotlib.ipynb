```python
import matplotlib.pyplot as plt
import numpy as np
```

###1. Basic Line Plot
####o Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
####o Label the x-axis as "X Values" and the y-axis as "Squared Values."
####o Add a title to the graph.

```python

x = np.linspace(-10, 10, 100)
y = x**2


plt.plot(x, y)
plt.xlabel('X Values')  # Label for x-axis
plt.ylabel('Squared Values')  # Label for y-axis
plt.title('Line Plot of y = x^2')  # Title of the graph

plt.show()
```

###2. Bar Plot
####o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi','Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
####o Add labels to the x-axis and y-axis and give the chart a title.

```python

students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]


plt.bar(students, marks, color='skyblue')
plt.xlabel('Students')  # Label for x-axis
plt.ylabel('Marks')  # Label for y-axis
plt.title('Marks Scored by Students')  # Title of the chart
plt.show()
```

####3. Scatter Plot
#####o Generate two lists of 20 random integers each (for x and y coordinates).
#####o Create a scatter plot with these values, and customize it by adding color,size, and labels to the points.

```python

import random


x_coords = [random.randint(1, 100) for _ in range(20)]
y_coords = [random.randint(1, 100) for _ in range(20)]

plt.scatter(x_coords, y_coords, c='green', s=100, alpha=0.6, edgecolors='black')
plt.xlabel('X Values')  # Label for x-axis
plt.ylabel('Y Values')  # Label for y-axis
plt.title('Scatter Plot with Random Values')  # Title of the plot
plt.show()
```

###4. Histogram
####o Simulate 100 random values from a normal distribution usingnp.random.randn() and plot a histogram.
####o Customize the number of bins and add a grid to the plot.

```python

random_values = np.random.randn(100)


plt.hist(random_values, bins=20, edgecolor='black')
plt.xlabel('Value')  # Label for x-axis
plt.ylabel('Frequency')  # Label for y-axis
plt.title('Histogram of Random Values')  # Title of the histogram
plt.grid(True)
plt.show()
```

###5. Pie Chart
####o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']with values [20, 30, 25, 15, 10].
####o Display the percentage values on the chart and use a color palette for differentiation.


```python

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]


plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Percentage Distribution of Categories')  # Title of the pie chart
plt.show()
```

###8. Comparison of Matplotlib and Seaborn
####o Generate a pair of plots (bar charts or line plots) using the same data inboth Matplotlib and Seaborn.
####o Highlight the differences in the code, syntax, and aesthetics of the twolibraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data = {'Category': ['A', 'B', 'C', 'D', 'E'], 'Value': [23, 45, 56, 78, 33]}
df = pd.DataFrame(data)

# Matplotlib Bar Plot
plt.bar(df['Category'], df['Value'], color='blue')
plt.title('Bar Chart using Matplotlib')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Seaborn Bar Plot
sns.barplot(x='Category', y='Value', data=df, color='green')
plt.title('Bar Chart using Seaborn')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()


```

###differences:
Matplotlib: Provides more control over plot details (color, size, labels), but requires more code and effort to make plots look good. It's great for customizing every aspect of the plot.

Seaborn: Simplifies plot creation with better default visuals and color schemes. It's easier to use for attractive plots but offers less flexibility for fine-tuning compared to Matplotlib.

###9. Pair Plot (Seaborn)

###10. Save and Style Plots
‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.style.use('ggplot')


plt.plot(x, y, label='Cosine Wave')
plt.title('Cosine Wave Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()


plt.savefig('cosine_wave_plot.png')
plt.savefig('cosine_wave_plot.pdf')

plt.show()
```
